```python
vectorizer = TfidfVectorizer(analyzer='word')
```

```python
pray_cuts={}
for name,pray in pray_docs.items():
    pray_cuts[name]=' '.join(jieba.cut(pray, cut_all=True))
```

```python
tfidf=vectorizer.fit_transform(list(pray_cuts.values()))
```

```python
from sklearn.metrics.pairwise import linear_kernel
cosine_similarities = linear_kernel(tfidf, tfidf).flatten()
```

```python
result=cosine_similarities.reshape(15,15)
```

```python
import pandas as pd
df=pd.DataFrame(result,columns=pray_docs.keys(),index=pray_docs.keys())
```
